```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv("ACC.csv")
df.head()
```

```python
initial_timestamp = df.columns[0]
frequency = df.iloc[0][initial_timestamp]
df = df.drop(0)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df_len = df.shape[0]
timestamp = []
for i in range(0,df_len):
    current_time = float(initial_timestamp)+i/frequency
    date_time = datetime.datetime.fromtimestamp(current_time).strftime('%d-%m-%Y %H:%M:%S:%f')
    timestamp.append(date_time)
    
```

```python
acc_df = pd.DataFrame()
acc_df['timestamp'] = timestamp
acc_df['x'] = df[df.columns[0]]
acc_df['y'] = df[df.columns[1]]
acc_df['z'] = df[df.columns[2]]
acc_df
```
