```python
import pandas as pd
import datetime
```

```python
acc = pd.read_csv("ACC.csv")
eda = pd.read_csv("EDA.csv")
temp = pd.read_csv("TEMP.csv")
hr = pd.read_csv("HR.csv")
```

```python
acc_timestamp = acc.columns[0]

acc_freq = acc.iloc[0][acc_timestamp]
eda_freq = eda.iloc[0][eda_timestamp]
temp_freq = temp.iloc[0][temp_timestamp]
hr_freq = hr.iloc[0][hr_timestamp]

acc = acc.drop(0)
acc.reset_index(drop=True, inplace=True)

eda = eda.drop(0)
eda.reset_index(drop=True, inplace=True)

temp = temp.drop(0)
temp.reset_index(drop=True, inplace=True)

hr = hr.drop(0)
hr.reset_index(drop=True, inplace=True)
```

```python
def calculate_timestamp(df,df_freq):
    df_len = df.shape[0]
    df_timestamp = df.columns[0]
    
    timestamp = {}
    for i in range(0,df_len):
        current_time = float(df_timestamp)+i/df_freq
        timestamp[str(current_time)] = df.iloc[i][0]
    return timestamp
```

```python
acc_len = acc.shape[0]
timestamp = []
for i in range(0,acc_len):
    current_time = float(acc_timestamp)+i/acc_freq
    timestamp.append(str(current_time))
```

```python
df = pd.DataFrame()
df['timestamp'] = timestamp
df['acc_x'] = acc[acc.columns[0]]
df['acc_y'] = acc[acc.columns[1]]
df['acc_z'] = acc[acc.columns[2]]
df['HR'] = df['timestamp'].map(calculate_timestamp(hr,hr_freq))
df['TEMP'] = df['timestamp'].map(calculate_timestamp(temp,temp_freq))
df['EDA'] = df['timestamp'].map(calculate_timestamp(eda,eda_freq))
df
```
